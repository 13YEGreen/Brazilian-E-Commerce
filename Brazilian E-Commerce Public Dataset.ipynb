```python
orders_2["order_months"] = orders_2["order_purchase_timestamp"].dt.to_period('M')
orders_2["cohort_months"] = orders_2.groupby("customer_unique_id", as_index = False)["order_months"].transform("min")
```

```python
orders_2[["customer_unique_id", "order_months", "cohort_months"]]
```

```python
order_year = orders_2["order_months"].dt.year
order_months = orders_2["order_months"].dt.month
cohort_year = orders_2["cohort_months"].dt.year
cohort_months = orders_2["cohort_months"].dt.month
year_difference = order_year - cohort_year
months_difference = order_months - cohort_months
```

```python
order_months
```

```python
months_difference
```

```python
orders_2["cohort_index"] = year_difference * 12 + months_difference
```

```python
orders_2["cohort_index"]
```

```python
cohort_unique_customer = orders_2.groupby(["cohort_months", "cohort_index"], as_index = False).agg({"customer_id":"nunique"})
cohort_unique_customer
```

```python
cohort_pivot = cohort_unique_customer.pivot(index = "cohort_months", values = "customer_id", columns = "cohort_index").fillna(0)
cohort_pivot
```

```python
cohort_per = cohort_pivot.iloc[:, 0]
cohort_per
```

```python
cohort_retention = cohort_pivot.divide(cohort_per, axis = 0)*100
display(cohort_retention.style.format('{:.2f}%'))
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data = cohort_retention/100, annot = True, fmt = '.0%', cmap = 'YlGnBu', linecolor = 'white', cbar = False, 
            annot_kws = {"size":12})
plt.title("Müştəri loyallığı", fontsize = 16, fontweight = 'bold') 
plt.show()
```
